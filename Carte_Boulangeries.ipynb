```python
import pandas as pd

# 🔧 1. Chargement de tes deux DataFrames (déjà fait dans ton cas)
# df = pd.read_csv("DS_BPE_data.csv", sep=';', encoding='utf-8')
# df_meta = pd.read_csv("Metadonnees.csv", sep=';', encoding='utf-8')

# ✅ 2. Première jointure : FACILITY_DOM
df = df.merge(
    df_meta[['COD_MOD', 'LIB_MOD']],
    left_on='FACILITY_DOM',
    right_on='COD_MOD',
    how='left'
)

df = df.rename(columns={'LIB_MOD': 'LIB_INSTAL'})
df = df.drop(columns=['COD_MOD'])  # Optionnel : nettoyer

# ✅ 3. Deuxième jointure : FACILITY_SDOM
df = df.merge(
    df_meta[['COD_MOD', 'LIB_MOD']],
    left_on='FACILITY_SDOM',
    right_on='COD_MOD',
    how='left'
)

df = df.rename(columns={'LIB_MOD': 'LIB_SS_INSTAL'})
df = df.drop(columns=['COD_MOD'])

# ✅ 4. Troisième jointure : FACILITY_TYPE
df = df.merge(
    df_meta[['COD_MOD', 'LIB_MOD']],
    left_on='FACILITY_TYPE',
    right_on='COD_MOD',
    how='left'
)

df = df.rename(columns={'LIB_MOD': 'LIB_TYP_INSTAL'})
df = df.drop(columns=['COD_MOD'])

# 🔎 5. Vérification finale
print(df.head())
print(df.columns)
```

```python
# 🔧 Sélectionner et réorganiser les colonnes souhaitées
df_clean = df[['GEO', 'OBS_VALUE', 'LIB_INSTAL', 'LIB_SS_INSTAL', 'LIB_TYP_INSTAL']].copy()

# 🔎 Vérification
print(df_clean.head())
print(df_clean.columns)
```

```python
# 🔧 Renommer les colonnes
df_clean = df_clean.rename(columns={
    'GEO': 'CODE_POSTAL',
    'OBS_VALUE': 'NB',
    'LIB_TYP_INSTAL': 'TYPE_INSTALLATION'
})

# 🔎 Vérification
print(df_clean.head())
print(df_clean.columns)
```

```python
# 🔧 Vérifier que CODE_POSTAL contient toujours 5 caractères

# Créer une colonne temporaire de longueur pour inspection rapide
df_clean['LENGTH_CP'] = df_clean['CODE_POSTAL'].astype(str).str.len()

# 🔎 Filtrer les lignes où la longueur diffère de 5
invalid_cp = df_clean[df_clean['LENGTH_CP'] != 5]

# Afficher le résultat
print(f"Nombre de lignes avec CODE_POSTAL != 5 caractères : {len(invalid_cp)}")
print(invalid_cp[['CODE_POSTAL']].drop_duplicates())
```

```python
# Compter le nombre de caractères distincts
print(df_clean['CODE_POSTAL'].astype(str).str.len().value_counts().sort_index())

# Afficher un échantillon de valeurs pour chaque longueur
for length in sorted(df_clean['CODE_POSTAL'].astype(str).str.len().unique()):
    subset = df_clean[df_clean['CODE_POSTAL'].astype(str).str.len() == length]
    print(f"\nLongueur {length} - exemples :")
    print(subset['CODE_POSTAL'].drop_duplicates().head(10).to_list())

```

```python
# 🔧 Filtrer sur longueurs utiles
df_clean['CODE_POSTAL'] = df_clean['CODE_POSTAL'].astype(str)

# Conserver uniquement longueurs 2, 3, 5
df_clean = df_clean[df_clean['CODE_POSTAL'].str.len().isin([2,3,5])].copy()

# Créer DEPARTEMENT
def extract_departement(code):
    if len(code) in [2,3]:
        return code.zfill(2)  # assure un format à 2 ou 3 digits avec padding si nécessaire
    elif len(code) == 5:
        return code[:2]  # pour commune, prendre les deux premiers chiffres
    else:
        return None

df_clean['DEPARTEMENT'] = df_clean['CODE_POSTAL'].apply(extract_departement)

# 🔎 Vérifier
print(df_clean[['CODE_POSTAL', 'DEPARTEMENT']].head())

```
